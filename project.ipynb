```python
import pandas as pd
a=pd.read_csv("/content/HepatitisCdata.csv")
print(a)
```

```python
a.drop(labels='Unnamed: 0',axis=1,inplace=True)
a
```

```python
y=a['Category']
print(y)
```

```python
x=a.drop('Category',axis=1)
x
```

```python
x['Sex'] = x['Sex'].replace({'m': 1, 'f' : 0})
x
```

```python
y = y.replace({'0=Blood Donor':0,'0s=suspect Blood Donor':1,'1=Hepatitis':2,'2=Fibrosis':3,'3=Cirrhosis':4})
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
model=SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn import svm
svm_model = svm.SVC(kernel='linear')
svm_model.fit(x_train, y_train)
y_pred = svm_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.linear_model import Perceptron
perceptron_model = Perceptron()
perceptron_model.fit(x_train, y_train)
y_pred = perceptron_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
